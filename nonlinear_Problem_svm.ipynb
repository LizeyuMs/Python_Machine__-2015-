```python
import numpy as np
import matplotlib.pyplot as plt
np.random.seed(0)
```

```python
X_xor = np.random.randn(200, 2)
y_xor = np.logical_xor(X_xor[:, 0] > 0,
                       X_xor[:, 1] > 0)
plt.scatter(X_xor[:,0], X_xor[:,1])
plt.show()
y_xor[:10]
```

```python
y_xor = np.where(y_xor==True, 1, -1)
plt.scatter(X_xor[y_xor==1, 0], X_xor[y_xor==1, 1],
            c='blue', marker='x', label='l')
plt.scatter(X_xor[y_xor==-1, 0], X_xor[y_xor==-1, 1],
            c='red', marker='o', label='-l')
plt.legend()
plt.show()
```
